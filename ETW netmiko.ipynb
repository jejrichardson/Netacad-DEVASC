```python
import datetime
print ("Current date and time: ")
print(datetime.datetime.now())
print("Configuring loopback interface and showing interface status")
#
from netmiko import ConnectHandler
sshCli = ConnectHandler(
    device_type="cisco_ios",
    host="192.168.56.107",
    port="22",
    username="cisco",
    password="cisco123!"
    )
config_commands = (
    'interface loopback 1' ,
    'ip address 10.1.1.1 255.255.255.0',
    'description LO1'
    )

output=sshCli.send_config_set(config_commands)

output=sshCli.send_command("show ip int br")

print(output)

```

```python
import datetime
print ("Current date and time: ")
print(datetime.datetime.now())
print("Configuring loopback interface ipv4 and ipv6 and showing interface status")
#
from netmiko import ConnectHandler
sshCli = ConnectHandler(
    device_type="cisco_ios",
    host="192.168.56.107",
    port="22",
    username="cisco",
    password="cisco123!"
    )
config_commands = (
    'interface loopback 2' ,
    'ip address 10.2.2.2 255.255.255.0',
    'ipv6 address 2001:db8:0:0:0:0:2:2/64',
    'description LO2'
    )
output=sshCli.send_config_set(config_commands)
#
output=sshCli.send_command("show ipv6 int br")
print(output)
```

```python
import datetime
print ("Current date and time: ")
print(datetime.datetime.now())
print("Removing loopback interface and showing interface status")
#
from netmiko import ConnectHandler
sshCli = ConnectHandler(
    device_type="cisco_ios",
    host="192.168.56.107",
    port="22",
    username="cisco",
    password="cisco123!"
    )
config_commands = (
    'int loopback 2' ,
    'no ip address',
    'exit',
    'no interface loopback 2'
    )
output=sshCli.send_config_set(config_commands)
output=sshCli.send_command("show ip int br")
print(output)
```
