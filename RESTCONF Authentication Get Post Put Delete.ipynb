```python
### RESTCONF IETF-INTERFACES
import requests
import json
import datetime
print ("Current date and time: ")
print(datetime.datetime.now())

# Disable SSL Verification Warning because of Private SSL Certificate
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

print('Creating PUT request for additional interface')
print('-------1-------')
### check ip address of local virtual router
router_ip = "192.168.56.107"
new_interface ='Loopback10111'
new_ip = '10.1.1.1'
print("Router IP: " +router_ip)

print('-------2-------')
api_url = "https://{0}/restconf/data/ietf-interfaces:interfaces/interface={1}".format(router_ip,new_interface)
print("Restconf URL: " +api_url)

print('-------3-------')
# A mistake in the headers generates an HTTP ERROR 406 Not Acceptable or HTTP 400 Malformed message
headers = {
    "Accept": "application/yang-data+json",
    "Content-type": "application/yang-data+json"
}
print(type(headers))
print(headers)

print('-------4-------')
basicauth =('cisco', 'cisco123!')
print(type(basicauth))
print(basicauth)

print('-------5-------')
yang_config = {
  "ietf-interfaces:interface": {
    "name": new_interface,
    "description": "New RESTCONF loopback",
    "type": "iana-if-type:softwareLoopback",
    "enabled": True,
    "ietf-ip:ipv4": {
      "address": [
        {
          "ip": new_ip,
          "netmask": "255.255.255.0"
        }
      ]
    },
    "ietf-ip:ipv6": {}
  }
}
print(type(yang_config))
print(yang_config)

print('-------6-------')
print(json.dumps(yang_config, indent=4))
print('-------7-------')
print("=> request executed in next cell")
```

```python
### RESTCONF: Delete interface
import requests
import json
import datetime
print ("Current date and time: ")
print(datetime.datetime.now())

# Disable SSL Verification Warning because of Private SSL Certificate
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

print('Sending DELETE request to remove interface')
del_interface = "Loopback" + input("What loopback number to delete? ")
print(del_interface)
### https://192.168.56.107/restconf/data/ietf-interfaces:interfaces/interface=Loopback10111
api_url = "https://{0}/restconf/data/ietf-interfaces:interfaces/interface={1}".format(router_ip, del_interface)

print('-------1-------')
yang_config = {
  "ietf-interfaces:interface": {
    "name": new_interface,
    "description": "",
    "type": "iana-if-type:softwareLoopback",
    "enabled": False,
    "ietf-ip:ipv4": {
      "address": [
        {
          "ip": "",
          "netmask": ""
        }
      ]
    },
    "ietf-ip:ipv6": {}
  }
}
resp = requests.request('DELETE', api_url, auth=basicauth, headers=headers, json=yang_config, verify=False)
print(resp.status_code)

print('-------2-------')
print(api_url)
print(resp.status_code)

print('-------3-------')
print(type(resp))
print(resp.text)
# SHOULD BE: 204 No Content
```

```python
### RESTCONF CISCO IOS-XE-INTERFACES
# https://github.com/jeremycohoe/restconf-catalyst-3850/blob/master/restconf_c3850.py
# Jeremy Cohoe jeremycohoe
# https://www.ciscolive.com/c/dam/r/ciscolive/us/docs/2019/pdf/5eU6DfQV/LTRCRT-2700-LG.pdf
# Tony Roman, LTRCRT-2700
import requests
from requests.auth import HTTPBasicAuth
import json
import datetime
print ("Current date and time: ")
print(datetime.datetime.now())

# Disable SSL Verification Warning because of Private SSL Certificate
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

auth = HTTPBasicAuth('cisco', 'cisco123!')
#print(dir(auth))
print(auth.username)
print(auth.password)
print(auth.__dict__)

media_type = 'application/yang-data+json'
headers = { 'Accept': media_type, 'Content-Type': media_type }
payload  = {}

url = "https://192.168.56.107/restconf/data/Cisco-IOS-XE-interfaces-oper:interfaces/interface?fields=name;speed;phys-address;admin-status;statistics/in-octets;statistics/out-octets"

response = requests.request("GET", url, auth=auth, headers=headers, data = payload, verify=False)

print("API: ", url)
print(response.status_code)

if response.status_code in [200, 202, 204]:
    print("Successful")
else: 
    print("Error in API")
    
#print(response.text.encode('utf8'))
print(response.json())
```
